# ODI to Databricks Migration: `trg_emp` Load

**Conversion Timestamp:** 2024-07-30 12:00:00

This notebook performs a full load into the `trg_emp` table from the `employees` table.

## Data Load

```sql
%sql
-- SCEN_TASK_NO in {10}
-- SCEN_TASK_NO in {20}
-- SCEN_TASK_NO in {30}
INSERT 
  INTO workspace.hr.trg_emp
  (
    employee_id ,
    first_name ,
    last_name ,
    email ,
    phone_number ,
    hire_date ,
    job_id ,
    salary ,
    commission_pct ,
    manager_id ,
    department_id 
  ) 
SELECT 
  employees.employee_id ,
  employees.first_name ,
  employees.last_name ,
  employees.email ,
  employees.phone_number ,
  employees.hire_date ,
  employees.job_id ,
  employees.salary ,
  employees.commission_pct ,
  employees.manager_id ,
  employees.department_id  
FROM 
  workspace.hr.employees AS employees;
```

## Validation

## Conversion Notes

*   The Oracle hint `/*+ APPEND PARALLEL */` has been removed as it is not applicable in Databricks Spark SQL.
*   Schema and table names (`HR.TRG_EMP`, `HR.EMPLOYEES`) have been converted to lowercase and prefixed with `workspace.` (`workspace.hr.trg_emp`, `workspace.hr.employees`) as per naming conventions.
*   Since no ODI parameters were present, no `dbutils.widgets` or temporary views for parameters were created.

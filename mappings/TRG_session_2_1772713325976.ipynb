# Conversion of TARGET.sql to Databricks Spark SQL

**Source File:** `TARGET.sql`
**Conversion Date:** 2024-07-30 12:00:00
**Description:** This notebook converts a simple Oracle INSERT statement into Databricks Spark SQL. It loads data from `HR.DEPARTMENTS` into `HR.TRG_DEP`.

```python
dbutils.widgets.text("ETL_JOB_TYPE", "FULL", "ETL Job Type (e.g., FULL, INCREMENTAL)")
dbutils.widgets.text("DATASOURCE_NUM_ID", "1", "Datasource Number ID")
dbutils.widgets.text("ETL_PROC_WID", "-1", "ETL Process Widget ID")
dbutils.widgets.text("ODI_SESS_NO", "-1", "ODI Session Number")
```

# ETL Parameters

```python
display(spark.sql("""
  SELECT
    '${ETL_JOB_TYPE}' AS ETL_JOB_TYPE,
    '${DATASOURCE_NUM_ID}' AS DATASOURCE_NUM_ID,
    '${ETL_PROC_WID}' AS ETL_PROC_WID,
    '${ODI_SESS_NO}' AS ODI_SESS_NO
"""))
```

# Insert into Target

```python
%sql
-- SCEN_TASK_NO in {10}
-- SCEN_TASK_NO in {20}
-- SCEN_TASK_NO in {30}

INSERT
  INTO workspace.hr.trg_dep
  (
    DEPARTMENT_ID ,
    DEPARTMENT_NAME ,
    MANAGER_ID ,
    LOCATION_ID
  )
SELECT
  DEPARTMENTS.DEPARTMENT_ID ,
  DEPARTMENTS.DEPARTMENT_NAME ,
  DEPARTMENTS.MANAGER_ID ,
  DEPARTMENTS.LOCATION_ID
FROM
  workspace.hr.departments DEPARTMENTS;
```

# Validation

```python
%sql
SELECT COUNT(*) AS target_record_count FROM workspace.hr.trg_dep;
```

# Conversion Notes

1.  **Oracle Hints Removed:** The Oracle `/*+ APPEND PARALLEL */` hint has been removed as it is not applicable in Databricks Spark SQL.
2.  **Schema and Table Naming:** Oracle schema `HR` has been converted to `workspace.hr`. Table names `TRG_DEP` and `DEPARTMENTS` have been converted to lowercase `trg_dep` and `departments` respectively, following Databricks naming conventions.
3.  **SCEN_TASK_NO:** The `SCEN_TASK_NO` entries have been preserved as comments within the relevant SQL cell.
4.  **ETL Parameters:** Standard ETL widgets have been created at the beginning of the notebook, although no specific ODI parameters (`#GLOBAL` or `#SCHEMA`) were found or used in the provided source SQL. These widgets provide a consistent interface for general ETL process control.

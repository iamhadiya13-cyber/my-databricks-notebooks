## Staging and Target Table DDL

```python
%sql
CREATE TABLE IF NOT EXISTS workspace.hr.trg_dep (
    DEPARTMENT_ID   BIGINT,
    DEPARTMENT_NAME STRING,
    MANAGER_ID      BIGINT,
    LOCATION_ID     BIGINT
) USING DELTA;
```

## Data Insertion

```python
%sql
-- SCEN_TASK_NO in {10}
-- SCEN_TASK_NO in {20}
-- SCEN_TASK_NO in {30}
INSERT INTO workspace.hr.trg_dep (
    DEPARTMENT_ID,
    DEPARTMENT_NAME,
    MANAGER_ID,
    LOCATION_ID
)
SELECT
    departments.DEPARTMENT_ID,
    departments.DEPARTMENT_NAME,
    departments.MANAGER_ID,
    departments.LOCATION_ID
FROM
    workspace.hr.departments AS departments;
```

- Data types for the `trg_dep` table were inferred from typical Oracle HR schema usage and mapped to Spark SQL equivalents (`NUMBER` to `BIGINT`, `VARCHAR2` to `STRING`). A `CREATE TABLE IF NOT EXISTS` statement was added for the target table.
